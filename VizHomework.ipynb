## Import Data into Data Frame

```python
import pandas as pd
df = pd.read_csv("Top_Highest_Openings.csv")
df.head()
```

```python

from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
```

## Graph 1: Histogram

```python
plt.hist(df['Average'], bins=10, color='purple', edgecolor='black')
plt.title('Histogram of Average Earnings')
plt.xlabel('Average Earnings Per Theater (Opening Weekend)')
plt.ylabel('Frequency of Average Earnings')
plt.grid(True)
plt.show()
```

## Graph 2: Bar Graph

```python
distributor_gross = df.groupby('Distributor')['Total Gross'].sum().reset_index()

distributor_gross = distributor_gross.sort_values(by='Total Gross', ascending=False)

plt.bar(distributor_gross['Distributor'], distributor_gross['Total Gross'], color='green')
plt.title('Total Gross Earnings by Distributor')
plt.xlabel('Distributor')
plt.ylabel('Total Gross Earnings (in Hundred Millions $)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

## Graph 3: Scatterplot

```python
plt.scatter(df['Opening'], df['Total Gross'], color='red', edgecolor='darkred', alpha=0.1)
plt.title('Opening Gross vs. Total Gross (in Hundred Million $)')
plt.xlabel('Opening Gross')
plt.ylabel('Total Gross')
plt.grid(True)
plt.tight_layout()
plt.show()
```
